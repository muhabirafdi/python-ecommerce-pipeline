```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Membuat kelas untuk struktur data
class Order:
  def __init__(self, order_id, customer_name, order_date, total_amount):
    self.order_id = order_id            #atribut
    self.customer_name = customer_name  #atribut
    self.order_date = order_date        #atribut
    self.total_amount = total_amount    #atribut

#Method untuk menghitung pajak dengan input tambahan tarif pajak (tax_rate)
  def calculate_tax(self, tax_rate):
    return self.total_amount * tax_rate

#Method untuk menampilkan data order
  def display_order(self):
    print(f"ID: {self.order_id} | Nama: {self.customer_name} | Total : {self.total_amount}")

#Membuat kelas untuk mengelola semua data order
class OrderProcessor:
  def __init__(self):
    self.orders = []    #list kosong untuk menampung semua data order

#Method untuk menambah data order, input 'order' ditambahkan ke list self.orders
  def add_order(self, order):
    self.orders.append(order)

#Method untuk menghitung total pendapatan dari semua data order
  def calculate_total_revenue(self):
    total = 0
    for x in self.orders:     #Loop untuk mengecek tiap pendapatan
      total += x.total_amount #Tiap pendapatan dijumlahkan dengan total pendapatan sebelumnya
    return total

#Method untuk menghitung total pajak dari semua data order
  def calculate_total_tax(self, tr):
    total_tax = 0
    for x in self.orders:                     #Loop untuk mengecek pajak tiap data order
      total_tax += Order.calculate_tax(x,tr)  #Tiap pajak dijumlahkan dengan total pajak sebelumnya
    return total_tax

#Method untuk menampilkan semua data order
  def display_orders(self):
    for x in self.orders:       #Loop untuk mengecek data order
      Order.display_order(x)    #Memanggil Method di kelas 'Order' untuk mencetak tiap data order

#Saya memasukkan 3 data order dengan total pendapatan 360000 dan total pajak 10% sebesar 36000

#Membuat 3 objek (pesanan1, pesanan2, pesanan3) untuk kelas 'Order'
pesanan1 = Order("001", "Budi", "2024-01-01", 100000)
pesanan2 = Order("001", "Budi", "2024-01-01", 120000)
pesanan3 = Order("001", "Budi", "2024-01-01", 140000)
#Input 'tax_rate' untuk menghitung total pajak
pajak = 0.1
#Objek yang sudah dibuat dimasukkan ke dalam list
prosesor = OrderProcessor()
prosesor.add_order(pesanan1)
prosesor.add_order(pesanan2)
prosesor.add_order(pesanan3)
#Menampilkan semua data order
prosesor.display_orders()
print(f"Total Pendapatan : {prosesor.calculate_total_revenue()}") #Menampilkan total pendapatan
print(f"Total Pajak : {prosesor.calculate_total_tax(pajak)}")     #Menampilkan total pajak

```
